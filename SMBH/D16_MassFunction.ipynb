```python
import numpy as np 
import matplotlib.pyplot as plt
from astropy.io import fits
from astropy.table import Table
from astropy.units import cds                   # adding the units to the values           
cds.enable()
from astropy import units as u
from astropy.cosmology import WMAP9 as cosmo  

DR16=fits.open("DR16_Z.fits")         #open function reads the file 
FA= Table(DR16[1].data)  
```

```python
Z=np.array(FA['Z']) 
M=np.array(FA['LOGMBH'])
name=np.array(FA['SDSS_NAME_1'])
RA=np.array(FA['RA_1'])
DEC=np.array(FA['DEC_1'])
LB=np.array(FA['LOGLBOL'])
UF=np.array(FA['FUV'])
RF=np.array(FA['FIRST_FLUX'])
XF=np.array(FA['XMM_TOTAL_FLUX'])
XFF=XF[XF>0]

from astropy.cosmology import WMAP9 as cosmo      #Distance Luminosity from redshift 
DLOX= cosmo.luminosity_distance(Z) 
from astropy import units as u                   #unit conversion from Megaparsec to cm 
DLLOX = (DLOX).to(u.cm)
DL=DLLOX.value
LX = 4 *np.pi * (DL**2) *XF
```
